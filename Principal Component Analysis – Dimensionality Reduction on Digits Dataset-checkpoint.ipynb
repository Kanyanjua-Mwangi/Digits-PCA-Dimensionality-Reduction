## Principal Component Analysis – Dimensionality Reduction on Digits Dataset

### Project Overview
The Digits dataset comprises 1,797 samples of 8×8 grayscale handwritten digit images (0–9), flattened into 64-dimensional feature vectors. High dimensionality in image data often leads to:
- Increased computational complexity
- Risk of overfitting in machine learning models
- Challenges in visualization and exploratory analysis

Principal Component Analysis (PCA) is applied to:
- Reduce the feature space while retaining the majority of the data variance
- Quantify the amount of information preserved by each principal component
- Identify the smallest number of components required to explain at least 85% of the total variance

This notebook demonstrates a standard PCA workflow for high-dimensional image data and provides quantitative justification for dimensionality reduction.

## Import libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

## Load and prepare dataset

```python
# Load the dataset
digits = load_digits()
X = digits.data
y = digits.target

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Dimensionality Reduction via PCA Transformation – Digits Dataset

### Objective
Transform the standardized 64-dimensional Digits dataset into a lower-dimensional representation using Principal Component Analysis (PCA). This step projects the data onto the principal components that capture the maximum variance, enabling efficient downstream analysis, visualization, and modeling while minimizing information loss.

### Implementation Code

```python
# Apply PCA without specifying number of components
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
print("PCA Transformation Results")
print("-" * 60)
print(f"Original shape:           {X_scaled.shape}")
print(f"Transformed shape (PCA):  {X_pca.shape}")
print(f"Number of components computed: {pca.n_components_}")
print(f"First sample (first 5 values after PCA):")
print(X_pca[0, :5].round(6))
print("-" * 60)
```

**Interpretation**:

- Original → Transformed shape: **(1797, 64)** → full PCA projection (no reduction yet)
- All **64 components** computed — complete basis with 100% variance preserved
- First sample PC scores: [-1.914, -0.955, -3.946, -2.029, -0.267]  
  → Strong loadings on PC1 and PC3, moderate on PC2/PC4

**Key Points**:
- Data now in **uncorrelated**, **variance-ordered** space
- First components explain most information
- Ready for reduction: select top k components (e.g., for ≥85% variance)

## Explained Variance Ratio – Principal Components (Digits Dataset)

**Objective**  
Quantify the proportion of total variance captured by each principal component. This metric indicates the relative importance of each PC and guides dimensionality reduction decisions.

**Implementation Code**

```python
explained_variance_ratio = pca.explained_variance_ratio_
for i, var in enumerate(explained_variance_ratio, start=1):
    print(f"PC{i}: {var:.4f}")
```

**Interpretation**:

- **PC1** explains the largest portion: **12.03%** of total variance  
- **PC2** follows with **9.56%**  
- Variance decreases steadily — **first 10 components** capture ~**55–60%** cumulatively  
- By **PC20**, ~**70–75%** variance is explained  
- Later components (PC50+) contribute very little (<0.3% each) — near-zero by PC62–64

**Key Insights**:

- Strong concentration in early components — typical for image data with correlated pixels
- First few PCs capture major structural patterns (e.g., stroke thickness, orientation)
- Rapid decay after ~PC20 indicates high redundancy in the 64-dimensional pixel space

**Implications**:

- Significant dimensionality reduction is possible — retaining **10–20 components** covers most information
- Components beyond ~PC30–40 add negligible value (noise or tiny details)

### Cumulative Explained Variance 

**Objective**  
Compute and display the cumulative explained variance ratio as principal components are added sequentially. This cumulative metric shows how much total variance is captured incrementally and helps determine the number of components needed to retain a substantial proportion of the original information (e.g., 85–95%).

**Implementation Code**

```python
cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

for i, var in enumerate(cumulative_variance_ratio, start=1):
    print(f"PC1–PC{i}: {var:.4f}")

cumulative_df = pd.DataFrame({
    "Principal Components": [f"PC1–PC{i}" for i in range(1, len(cumulative_variance_ratio) + 1)],
    "Cumulative Variance Ratio": [f"{v:.4f}" for v in cumulative_variance_ratio]
})

cumulative_df.head(10)
```

**Interpretation**:

- **Rapid early accumulation**: First 5 PCs capture ~41%, first 10 ~59%, first 15 ~71%
- **85% threshold reached**: At **PC25** (cumulative 0.8513 → 85.13%)
- **Diminishing returns**: After PC25–30, each additional component adds <1% variance
- **Near-full coverage**: By PC40 ~95%, by PC50 ~98%, full 100% at PC61 (last meaningful components; PC62–64 ≈0)

**Implications**:

- **Strong compression potential**: Retain **~25 components** for ≥85% variance (61% reduction from 64)
- **High redundancy** in pixel data: most information concentrated in first ~20–30 PCs
- **Noise in later components**: PCs beyond ~40–50 contribute negligible signal

**Conclusion**  
The cumulative variance curve shows that **~25 principal components** are sufficient to explain **>85%** of the total variance in the Digits dataset — confirming PCA's effectiveness for substantial dimensionality reduction with minimal information loss.
